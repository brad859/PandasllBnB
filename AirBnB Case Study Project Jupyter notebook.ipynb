```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('nycairbnb.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby(['neighbourhood_group']).count()[['neighbourhood']].sort_values('neighbourhood', ascending= False)
# Manhattan has the most listings
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby(['room_type', 'neighbourhood_group']).count()[['neighbourhood']].sort_values('room_type', ascending= False)
# No the entire home/apt is the most popular.
```

```python
# Which hosts are the busiest and based on their reviews?
air_bnb.groupby(['host_id', 'host_name']).count()[['number_of_reviews']].sort_values('number_of_reviews', ascending= False).head()
# Sonder (NYC) is the busiest based on reviews 
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby(['neighbourhood_group'])[['price']].mean().sort_values('price', ascending= False)
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby(['neighbourhood_group'])[['price']].max().sort_values('price', ascending= False)
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby(['host_id','host_name'])[['price']].max().sort_values('price', ascending= False).head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb[air_bnb['number_of_reviews'] > 100].sort_values('availability_365', ascending = True)[['host_id','host_name','number_of_reviews','availability_365']].head()
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.groupby(['host_id','host_name','neighbourhood','neighbourhood_group'])[['price']].max().sort_values('price', ascending= False).head()
```

```python
# When did Danielle from Queens last receive a review?
air_bnb[air_bnb['host_name'] == 'Danielle'].sort_values('last_review', ascending = False)[['host_id','host_name','neighbourhood_group','last_review']].head(5)
```
